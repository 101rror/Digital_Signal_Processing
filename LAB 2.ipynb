```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(-12, 13)
n = t
x = 2 * np.pi
```

```python
plt.figure()
f1 = 0
x1 = np.cos(x * f1 * n)
plt.stem(n, x1)
plt.xlabel('n')
plt.ylabel('x1')
plt.title('Figure 1')
plt.show()
```
