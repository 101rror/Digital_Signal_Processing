```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = int(input("Enter the length of Exponential sequence : "))
t = np.arange(0, n + 1, 1)
a = float(input("Enter the value of a : "))
y = np.exp(a*t)
```

```python
plt.stem(t, y)
plt.xlabel('(d)n------->')
plt.ylabel('Amplitude-------->')
plt.title('Exponential sequence')
plt.show()
```
